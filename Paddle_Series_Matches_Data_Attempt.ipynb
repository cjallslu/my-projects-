```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(urllib.request.urlopen("https://mycptc.org/node/98/matches").read(), 'lxml')
```

```python
tbody = soup('table', {"class":"views-table cols-6 table table-hover table-striped"})[0].find_all('tr')
```

```python
for row in tbody:
    cols = row.findChildren(recursive=False)
    print(cols)
```
